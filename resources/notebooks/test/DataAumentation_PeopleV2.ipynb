```python
# ## **Bloque 3: Configuraci√≥n**
class Config:
    EXPECTED_SPLITS = ['train', 'val', 'test']
    IMAGE_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.bmp'}
    REPORT_FILE = 'dataset_people_report.txt'
    PLOT_SIZE = (15, 8)
    CLASS_NAME = 'person'  # Clase esperada (clase 0)

# ## **Bloque 4: Funci√≥n para cargar configuraci√≥n YAML**
def load_yaml_config(base_path):
    yaml_path = base_path / 'data.yaml'
    if yaml_path.exists():
        with open(yaml_path) as f:
            return yaml.safe_load(f)
    return None

# ## **Bloque 5: Funci√≥n de validaci√≥n de estructura**
def validate_structure(base_path):
    missing = []
    required_dirs = ['images', 'labels']

    for dir in required_dirs:
        if not (base_path/dir).exists():
            missing.append(f"{dir}/")

    for split in Config.EXPECTED_SPLITS:
        if not (base_path/'images'/split).exists():
            missing.append(f"images/{split}/")
        if not (base_path/'labels'/split).exists():
            missing.append(f"labels/{split}/")

    if missing:
        print("\n‚ö†Ô∏è **Advertencia:** Directorios faltantes:")
        for item in missing:
            print(f"- {item}")
        print("\nEl an√°lisis podr√≠a ser incompleto.\n")
        return False
    return True

# ## **Bloque 6: Funci√≥n de validaci√≥n de bounding boxes**
def validate_bbox(bbox):
    if len(bbox) != 5:
        return False, "Formato incorrecto (deben ser 5 valores)"

    try:
        class_id, x_center, y_center, width, height = map(float, bbox)
        if class_id != 0:
            return False, f"Class ID inv√°lido: {class_id} (debe ser 0)"

        for val in [x_center, y_center, width, height]:
            if not (0 <= val <= 1):
                return False, f"Valor fuera de rango: {val} (debe estar entre 0-1)"

        return True, ""
    except ValueError:
        return False, "Valores no num√©ricos"

# ## **Bloque 7: Funci√≥n de an√°lisis del dataset**
def analyze_dataset(base_path):
    stats = {split: {
        'images': 0,
        'labels': 0,
        'missing_labels': [],
        'missing_images': [],
        'invalid_bboxes': [],
        'total_bboxes': 0,
        'class_distribution': defaultdict(int)
    } for split in Config.EXPECTED_SPLITS}

    for split in Config.EXPECTED_SPLITS:
        # Rutas de im√°genes y labels
        img_dir = base_path / 'images' / split
        label_dir = base_path / 'labels' / split

        # Procesar im√°genes
        images = list(img_dir.glob('*'))
        images = [f for f in images if f.suffix.lower() in Config.IMAGE_EXTENSIONS]
        stats[split]['images'] = len(images)

        # Procesar labels
        labels = list(label_dir.glob('*.txt'))
        stats[split]['labels'] = len(labels)

        # Verificar correspondencia
        for img in images:
            label_path = label_dir / (img.stem + '.txt')
            if not label_path.exists():
                stats[split]['missing_labels'].append(img.name)

        for lbl in labels:
            img_path = img_dir / (lbl.stem + '.jpg')
            if not img_path.exists():
                img_path = img_dir / (lbl.stem + '.png')
            if not img_path.exists():
                stats[split]['missing_images'].append(lbl.name)

        # Validar bounding boxes
        for lbl_file in tqdm(labels, desc=f"Validando {split}", unit="file"):
            try:
                with open(lbl_file) as f:
                    lines = f.readlines()

                for line_num, line in enumerate(lines, 1):
                    line = line.strip()
                    if not line:
                        continue

                    bbox = line.split()
                    is_valid, error = validate_bbox(bbox)
                    stats[split]['total_bboxes'] += 1
                    stats[split]['class_distribution'][int(bbox[0])] += 1

                    if not is_valid:
                        stats[split]['invalid_bboxes'].append({
                            'file': lbl_file.name,
                            'line': line_num,
                            'error': error,
                            'content': line
                        })
            except Exception as e:
                stats[split]['invalid_bboxes'].append({
                    'file': lbl_file.name,
                    'error': f"Error de lectura: {str(e)}"
                })

    return stats

# ## **Bloque 8: Funci√≥n de generaci√≥n de reporte**
def generate_report(stats, base_path):
    report = []
    report.append(f"# üìä Reporte de Dataset de Detecci√≥n de Personas\n")
    report.append(f"**Ruta del dataset:** `{base_path}`\n")

    # Resumen general
    report.append("## üìå Resumen General")
    summary_data = []
    for split in Config.EXPECTED_SPLITS:
        summary_data.append({
            'Split': split.upper(),
            'Im√°genes': stats[split]['images'],
            'Etiquetas': stats[split]['labels'],
            'BBoxes V√°lidos': stats[split]['total_bboxes'],
            'Labels Faltantes': len(stats[split]['missing_labels']),
            'Im√°genes Faltantes': len(stats[split]['missing_images']),
            'BBoxes Inv√°lidos': len(stats[split]['invalid_bboxes'])
        })

    df_summary = pd.DataFrame(summary_data)
    report.append(df_summary.to_markdown(index=False))

    # Problemas detectados
    report.append("\n## ‚ùå Problemas Detectados")
    for split in Config.EXPECTED_SPLITS:
        report.append(f"\n### {split.upper()}:")
        if stats[split]['missing_labels']:
            report.append(f"- Im√°genes sin labels: {len(stats[split]['missing_labels'])}")
        if stats[split]['missing_images']:
            report.append(f"- Labels sin im√°genes: {len(stats[split]['missing_images'])}")
        if stats[split]['invalid_bboxes']:
            report.append(f"- Bounding boxes inv√°lidos: {len(stats[split]['invalid_bboxes'])}")

    # Ejemplos de errores
    report.append("\n## üö® Ejemplos de Errores")
    for split in Config.EXPECTED_SPLITS:
        if stats[split]['invalid_bboxes']:
            report.append(f"\n### {split.upper()} - Primeros 3 errores:")
            for error in stats[split]['invalid_bboxes'][:3]:
                report.append(f"- **{error['file']}** (L√≠nea {error.get('line', '?')}):")
                report.append(f"  `{error.get('content', '')}`")
                report.append(f"  Error: {error['error']}")

    return '\n'.join(report)

# ## **Bloque 9: Visualizaci√≥n de datos**
def visualize_stats(stats):
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Gr√°fico 1: Distribuci√≥n de im√°genes
    splits = Config.EXPECTED_SPLITS
    image_counts = [stats[s]['images'] for s in splits]
    axes[0,0].bar(splits, image_counts)
    axes[0,0].set_title('Distribuci√≥n de Im√°genes por Split')

    # Gr√°fico 2: Bounding boxes por split
    bbox_counts = [stats[s]['total_bboxes'] for s in splits]
    axes[0,1].pie(bbox_counts, labels=splits, autopct='%1.1f%%')
    axes[0,1].set_title('Distribuci√≥n de Bounding Boxes')

    # Gr√°fico 3: Distribuci√≥n de clases
    class_counts = [stats[s]['class_distribution'].get(0, 0) for s in splits]
    axes[1,0].bar(splits, class_counts)
    axes[1,0].set_title('Detecciones de Personas (Clase 0)')

    # Gr√°fico 4: Problemas por split
    problems = {
        'Missing Labels': [len(stats[s]['missing_labels']) for s in splits],
        'Invalid BBoxes': [len(stats[s]['invalid_bboxes']) for s in splits]
    }
    df_problems = pd.DataFrame(problems, index=splits)
    df_problems.plot(kind='bar', ax=axes[1,1])
    axes[1,1].set_title('Problemas Detectados por Split')

    plt.tight_layout()
    plt.show()
```

```python
import os

def validate_dataset(images_dir, labels_dir):
    # Obtener lista de nombres de im√°genes (sin extensi√≥n)
    image_files = [f for f in os.listdir(images_dir) if os.path.isfile(os.path.join(images_dir, f))]
    image_names = {os.path.splitext(f)[0] for f in image_files}

    # Obtener lista de nombres de archivos .txt
    label_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt') and os.path.isfile(os.path.join(labels_dir, f))]
    label_names = {os.path.splitext(f)[0] for f in label_files}

    # Verificar im√°genes sin etiquetas
    missing_labels = image_names - label_names
    if missing_labels:
        print(f"\n--- Alerta: {len(missing_labels)} im√°genes sin archivos .txt correspondientes ---")
        for name in sorted(missing_labels):
            print(f"Imagen sin etiqueta: {name}")
    else:
        print("\nTodas las im√°genes tienen etiquetas correspondientes.")

    # Verificar etiquetas vac√≠as
    empty_labels = []
    for label_file in label_files:
        label_path = os.path.join(labels_dir, label_file)
        with open(label_path, 'r', encoding='utf-8') as f:
            content = f.read().strip()
            if not content:
                empty_labels.append(label_file)

    if empty_labels:
        print(f"\n--- Alerta: {len(empty_labels)} archivos .txt vac√≠os detectados ---")
        for label in sorted(empty_labels):
            print(f"Etiqueta vac√≠a: {label}")
    else:
        print("\nTodos los archivos .txt contienen datos.")

    # Resumen final
    print("\n--- Resumen de la validaci√≥n del dataset ---")
    print(f"Total de im√°genes: {len(image_names)}")
    print(f"Total de etiquetas: {len(label_names)}")
    print(f"Im√°genes sin etiquetas: {len(missing_labels)}")
    print(f"Etiquetas vac√≠as: {len(empty_labels)}")

    if not missing_labels and not empty_labels:
        print("\n¬°Validaci√≥n completada con √©xito! Todo est√° en orden.")
    else:
        print("\n¬°Validaci√≥n completada con advertencias! Revisa los problemas reportados.")

# Rutas de las carpetas (modifica estas variables seg√∫n tus necesidades)
images_dir = r'/content/drive/MyDrive/Proyecto IA-3/violence_school_project/dataset_people/images/test'  # Carpeta con las im√°genes
labels_dir = r'/content/drive/MyDrive/Proyecto IA-3/violence_school_project/dataset_people/labels/test'    # Carpeta con los archivos .txt de anotaciones

# Ejecutar la validaci√≥n
validate_dataset(images_dir, labels_dir)
```

```python
def visualize_bbox(img, bboxes, class_ids, class_name="person"):
    """Visualiza la imagen con sus bounding boxes para verificaci√≥n"""
    img = img.copy()
    h, w, _ = img.shape

    for i, bbox in enumerate(bboxes):
        # Convertir de formato YOLO [x_center, y_center, width, height] a p√≠xeles
        x_center, y_center, width, height = bbox
        x_min = int((x_center - width/2) * w)
        y_min = int((y_center - height/2) * h)
        x_max = int((x_center + width/2) * w)
        y_max = int((y_center + height/2) * h)

        # Dibujar rect√°ngulo
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

        # Etiqueta
        cv2.putText(img, class_name, (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    return img

def apply_augmentation(image_path, label_path, output_image_dir, output_label_dir, transform,
                       num_augmentations=3, visualize=False, viz_sample_prob=0.05):
    """Aplica aumentaci√≥n a una imagen y sus etiquetas"""
    # Leer imagen
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error al leer la imagen: {image_path}")
        return

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Leer etiquetas
    try:
        with open(label_path, 'r') as f:
            lines = f.readlines()
    except Exception as e:
        print(f"Error al leer el archivo de etiquetas {label_path}: {e}")
        return

    bboxes = []
    class_labels = []

    # Procesar etiquetas
    for line in lines:
        try:
            values = line.strip().split()
            class_id = int(values[0])
            x_center, y_center, width, height = [float(x) for x in values[1:5]]

            # Verificar si los valores est√°n en el rango correcto (0-1 para YOLO)
            if not all(0 <= val <= 1 for val in [x_center, y_center, width, height]):
                print(f"Advertencia: Valores fuera de rango en {label_path}: {values}")
                continue

            bboxes.append([x_center, y_center, width, height])
            class_labels.append(class_id)
        except Exception as e:
            print(f"Error al procesar l√≠nea en {label_path}: {line} - {e}")
            continue

    # Si no hay cajas v√°lidas, omitir esta imagen
    if not bboxes:
        print(f"No se encontraron bounding boxes v√°lidos en {label_path}")
        return

    # Obtener nombre base del archivo
    base_name = os.path.basename(image_path).split('.')[0]

    # Visualizar imagen original
    if visualize and random.random() < viz_sample_prob:
        vis_img = visualize_bbox(image, bboxes, class_labels)
        plt.figure(figsize=(10, 10))
        plt.imshow(vis_img)
        plt.title("Original Image")
        plt.axis('off')
        plt.savefig(os.path.join(VIZ_DIR, f"{base_name}_original.jpg"), dpi=100)
        plt.close()

    # Crear augmentaciones
    augmentation_attempts = 0
    successful_augmentations = 0

    # Intentaremos m√°s veces para conseguir el n√∫mero deseado de aumentaciones v√°lidas
    while successful_augmentations < num_augmentations and augmentation_attempts < num_augmentations * 2:
        augmentation_attempts += 1

        try:
            # Aplicar transformaci√≥n
            augmented = transform(image=image, bboxes=bboxes, class_labels=class_labels)

            aug_image = augmented['image']
            aug_bboxes = augmented['bboxes']
            aug_class_labels = augmented['class_labels']

            # Verificar que la aumentaci√≥n contiene los bounding boxes esperados
            if not aug_bboxes:
                continue

            # Verificar que los bounding boxes no est√°n demasiado distorsionados
            valid_boxes = True
            for bbox in aug_bboxes:
                _, _, w, h = bbox
                # Rechazar si el ancho o alto son demasiado peque√±os o grandes
                if w < 0.02 or h < 0.02 or w > 0.95 or h > 0.95:
                    valid_boxes = False
                    break

            if not valid_boxes:
                continue

            # Si llegamos aqu√≠, la aumentaci√≥n es v√°lida
            successful_augmentations += 1

            # Guardar imagen aumentada
            output_image_path = os.path.join(output_image_dir, f"{base_name}_aug{successful_augmentations}.jpg")
            cv2.imwrite(output_image_path, cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR))

            # Guardar etiquetas aumentadas
            output_label_path = os.path.join(output_label_dir, f"{base_name}_aug{successful_augmentations}.txt")
            with open(output_label_path, 'w') as f:
                for bbox, class_id in zip(aug_bboxes, aug_class_labels):
                    f.write(f"{class_id} {' '.join([str(x) for x in bbox])}\n")

            # Visualizar aumentaciones
            if visualize and random.random() < viz_sample_prob:
                vis_aug_img = visualize_bbox(aug_image, aug_bboxes, aug_class_labels)
                plt.figure(figsize=(10, 10))
                plt.imshow(vis_aug_img)
                plt.title(f"Augmented Image {successful_augmentations}")
                plt.axis('off')
                plt.savefig(os.path.join(VIZ_DIR, f"{base_name}_aug{successful_augmentations}.jpg"), dpi=100)
                plt.close()

        except Exception as e:
            print(f"Error en la augmentaci√≥n para {image_path}: {e}")
            continue

def process_split(split_name="train", augmentations_per_image=3, visualize_sample=True, max_images=None):
    """Procesa un conjunto de datos (train/val/test) con opci√≥n de limitar im√°genes para pruebas"""
    print(f"\n{'='*50}")
    print(f"Procesando split: {split_name}")
    print(f"{'='*50}")

    # Directorios originales
    src_image_dir = os.path.join(DATASET_ROOT, "images", split_name)
    src_label_dir = os.path.join(DATASET_ROOT, "labels", split_name)

    # Directorios de salida
    output_image_dir = os.path.join(OUTPUT_ROOT, "images", split_name)
    output_label_dir = os.path.join(OUTPUT_ROOT, "labels", split_name)

    # Seleccionar transformaci√≥n seg√∫n el split
    transform = train_transform if split_name == "train" else val_transform

    # Obtener todas las im√°genes
    image_files = glob.glob(os.path.join(src_image_dir, "*.jpg"))
    if not image_files:
        image_files = glob.glob(os.path.join(src_image_dir, "*.png"))

    if not image_files:
        print(f"ADVERTENCIA: No se encontraron im√°genes en {src_image_dir}")
        return

    print(f"Encontradas {len(image_files)} im√°genes en {src_image_dir}")

    # Limitar el n√∫mero de im√°genes para pruebas si es necesario
    if max_images is not None and max_images > 0 and max_images < len(image_files):
        print(f"Limitando procesamiento a {max_images} im√°genes (para pruebas)")
        random.shuffle(image_files)  # Aleatorizamos para tener una muestra representativa
        image_files = image_files[:max_images]

    # Copiar archivos originales primero
    print("Copiando archivos originales...")
    for img_path in tqdm(image_files):
        base_name = os.path.basename(img_path)
        label_name = os.path.splitext(base_name)[0] + ".txt"
        label_path = os.path.join(src_label_dir, label_name)

        if os.path.exists(label_path):
            try:
                shutil.copy(img_path, os.path.join(output_image_dir, base_name))
                shutil.copy(label_path, os.path.join(output_label_dir, label_name))
            except Exception as e:
                print(f"Error al copiar archivos originales: {e}")

    # Aplicar aumentaci√≥n solo si se requiere
    if augmentations_per_image > 0:
        print(f"Aplicando aumentaci√≥n ({augmentations_per_image} por imagen)...")
        for img_path in tqdm(image_files):
            base_name = os.path.basename(img_path)
            label_name = os.path.splitext(base_name)[0] + ".txt"
            label_path = os.path.join(src_label_dir, label_name)

            if os.path.exists(label_path):
                apply_augmentation(
                    img_path,
                    label_path,
                    output_image_dir,
                    output_label_dir,
                    transform,
                    num_augmentations=augmentations_per_image,
                    visualize=visualize_sample,
                    viz_sample_prob=0.05  # Visualizamos m√°s para asegurar
                )

def update_yaml_file():
    """Actualiza o crea el archivo data.yaml para el dataset aumentado"""
    src_yaml = os.path.join(DATASET_ROOT, "data.yaml")
    dst_yaml = os.path.join(OUTPUT_ROOT, "data.yaml")

    if os.path.exists(src_yaml):
        # Leer el YAML original
        with open(src_yaml, 'r') as f:
            yaml_content = f.read()

        # Actualizar rutas (para el caso de Drive, mejor usar rutas absolutas)
        updated_content = []
        for line in yaml_content.split('\n'):
            if line.startswith('train:') or line.startswith('val:') or line.startswith('test:'):
                key = line.split(':')[0].strip()
                updated_content.append(f"{key}: {os.path.join(OUTPUT_ROOT, 'images', key)}")
            else:
                updated_content.append(line)

        yaml_content = '\n'.join(updated_content)

        # Guardar nuevo YAML
        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml actualizado en {dst_yaml}")
    else:
        # Crear un YAML b√°sico
        train_path = os.path.join(OUTPUT_ROOT, "images", "train")
        val_path = os.path.join(OUTPUT_ROOT, "images", "val")
        test_path = os.path.join(OUTPUT_ROOT, "images", "test")

        yaml_content = f"""# Dataset configuraci√≥n para YOLOv11
train: {train_path}
val: {val_path}
test: {test_path}

# N√∫mero de clases
nc: 1

# Nombres de clases
names: ['person']
"""

        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml creado en {dst_yaml}")

def print_summary():
    """Imprime un resumen del proceso de aumentaci√≥n"""
    print("\n\n" + "="*60)
    print("RESUMEN DE AUMENTACI√ìN DE DATOS".center(60))
    print("="*60)

    for split in ["train", "val", "test"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        aug_img_dir = os.path.join(OUTPUT_ROOT, "images", split)

        if not os.path.exists(img_dir) or not os.path.exists(aug_img_dir):
            print(f"Split {split.upper()}: Directorio no encontrado, omitiendo...")
            continue

        orig_count = len(glob.glob(os.path.join(img_dir, "*.jpg")))
        if orig_count == 0:
            orig_count = len(glob.glob(os.path.join(img_dir, "*.png")))

        aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.jpg")))
        if aug_count == 0:
            aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.png")))

        if orig_count > 0:
            print(f"Split {split.upper()}:")
            print(f"  - Im√°genes originales: {orig_count}")
            print(f"  - Im√°genes despu√©s de aumentaci√≥n: {aug_count}")
            print(f"  - Incremento: {aug_count - orig_count} im√°genes ({(aug_count/orig_count - 1)*100:.1f}% m√°s)")
            print("-"*60)

    print("\nVisualizaciones guardadas en:", VIZ_DIR)
    print(f"\nDataset aumentado guardado en: {OUTPUT_ROOT}")
    print("="*60)

def test_run_single_image():
    """Ejecuta una prueba en una sola imagen para verificar que todo funciona"""
    print("\nEjecutando prueba en una sola imagen...")

    # Buscar una imagen de prueba
    for split in ["train", "val"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        if os.path.exists(img_dir):
            image_files = glob.glob(os.path.join(img_dir, "*.jpg"))
            if not image_files:
                image_files = glob.glob(os.path.join(img_dir, "*.png"))

            if image_files:
                test_img = image_files[0]
                base_name = os.path.basename(test_img)
                label_name = os.path.splitext(base_name)[0] + ".txt"
                label_path = os.path.join(DATASET_ROOT, "labels", split, label_name)

                if os.path.exists(label_path):
                    print(f"Probando con imagen: {test_img}")
                    output_img_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    output_label_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    os.makedirs(output_img_dir, exist_ok=True)

                    # Aplicar aumentaci√≥n a esta imagen solamente
                    transform = train_transform
                    apply_augmentation(
                        test_img,
                        label_path,
                        output_img_dir,
                        output_label_dir,
                        transform,
                        num_augmentations=5,  # Generamos 5 para ver variedad
                        visualize=True,
                        viz_sample_prob=1.0  # Siempre visualizar
                    )

                    print(f"Prueba completada. Verifica el directorio: {output_img_dir}")
                    return True

    print("No se pudo encontrar una imagen de prueba v√°lida.")
    return False
```

```python
def visualize_bbox(img, bboxes, class_ids, class_name="person"):
    """Visualiza la imagen con sus bounding boxes para verificaci√≥n"""
    img = img.copy()
    h, w, _ = img.shape

    for i, bbox in enumerate(bboxes):
        # Convertir de formato YOLO [x_center, y_center, width, height] a p√≠xeles
        x_center, y_center, width, height = bbox
        x_min = int((x_center - width/2) * w)
        y_min = int((y_center - height/2) * h)
        x_max = int((x_center + width/2) * w)
        y_max = int((y_center + height/2) * h)

        # Dibujar rect√°ngulo
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

        # Etiqueta
        cv2.putText(img, class_name, (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    return img

def apply_augmentation_type(image_path, label_path, output_image_dir, output_label_dir,
                            transform_type, aug_name, visualize=False):
    """Aplica un tipo espec√≠fico de transformaci√≥n a una imagen"""
    # Leer imagen
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error al leer la imagen: {image_path}")
        return False

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Leer etiquetas
    try:
        with open(label_path, 'r') as f:
            lines = f.readlines()
    except Exception as e:
        print(f"Error al leer el archivo de etiquetas {label_path}: {e}")
        return False

    bboxes = []
    class_labels = []

    # Procesar etiquetas
    for line in lines:
        try:
            values = line.strip().split()
            class_id = int(values[0])
            x_center, y_center, width, height = [float(x) for x in values[1:5]]

            # Verificar si los valores est√°n en el rango correcto (0-1 para YOLO)
            if not all(0 <= val <= 1 for val in [x_center, y_center, width, height]):
                print(f"Advertencia: Valores fuera de rango en {label_path}: {values}")
                continue

            bboxes.append([x_center, y_center, width, height])
            class_labels.append(class_id)
        except Exception as e:
            print(f"Error al procesar l√≠nea en {label_path}: {line} - {e}")
            continue

    # Si no hay cajas v√°lidas, omitir esta imagen
    if not bboxes:
        print(f"No se encontraron bounding boxes v√°lidos en {label_path}")
        return False

    # Obtener nombre base del archivo
    base_name = os.path.basename(image_path).split('.')[0]

    try:
        # Aplicar transformaci√≥n
        augmented = transform_type(image=image, bboxes=bboxes, class_labels=class_labels)

        aug_image = augmented['image']
        aug_bboxes = augmented['bboxes']
        aug_class_labels = augmented['class_labels']

        # Verificar que la aumentaci√≥n contiene los bounding boxes esperados
        if not aug_bboxes:
            print(f"La transformaci√≥n {aug_name} elimin√≥ todos los bounding boxes para {image_path}, omitiendo.")
            return False

        # Guardar imagen aumentada
        output_image_path = os.path.join(output_image_dir, f"{base_name}_aug_{aug_name}.jpg")
        cv2.imwrite(output_image_path, cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR))

        # Guardar etiquetas aumentadas
        output_label_path = os.path.join(output_label_dir, f"{base_name}_aug_{aug_name}.txt")
        with open(output_label_path, 'w') as f:
            for bbox, class_id in zip(aug_bboxes, aug_class_labels):
                f.write(f"{class_id} {' '.join([str(x) for x in bbox])}\n")

        # Visualizar aumentaci√≥n si se solicita
        if visualize:
            vis_aug_img = visualize_bbox(aug_image, aug_bboxes, aug_class_labels)
            plt.figure(figsize=(10, 10))
            plt.imshow(vis_aug_img)
            plt.title(f"Augmented Image - {aug_name}")
            plt.axis('off')
            plt.savefig(os.path.join(VIZ_DIR, f"{base_name}_aug_{aug_name}.jpg"), dpi=100)
            plt.close()

        return True  # La aumentaci√≥n se realiz√≥ correctamente

    except Exception as e:
        print(f"Error en la aumentaci√≥n {aug_name} para {image_path}: {e}")
        return False

def process_limited_split(split_name, total_images, visualize_sample=True):
    """
    Procesa una cantidad limitada de im√°genes para cada tipo de aumentaci√≥n.
    Garantiza que no se procese la misma imagen dos veces.

    Args:
        split_name: Nombre del split ('train' o 'val')
        total_images: N√∫mero total de im√°genes a procesar (se dividir√° entre los 2 tipos)
        visualize_sample: Si se deben visualizar algunas muestras
    """
    print(f"\n{'='*50}")
    print(f"Procesando split {split_name} con {total_images} im√°genes totales")
    print(f"{'='*50}")

    # Directorios originales
    src_image_dir = os.path.join(DATASET_ROOT, "images", split_name)
    src_label_dir = os.path.join(DATASET_ROOT, "labels", split_name)

    # Directorios de salida
    output_image_dir = os.path.join(OUTPUT_ROOT, "images", split_name)
    output_label_dir = os.path.join(OUTPUT_ROOT, "labels", split_name)

    # Obtener todas las im√°genes
    image_files = glob.glob(os.path.join(src_image_dir, "*.jpg"))
    if not image_files:
        image_files = glob.glob(os.path.join(src_image_dir, "*.png"))

    if not image_files:
        print(f"ADVERTENCIA: No se encontraron im√°genes en {src_image_dir}")
        return

    print(f"Encontradas {len(image_files)} im√°genes en {src_image_dir}")

    # Copiar archivos originales primero
    print("Copiando archivos originales...")
    valid_image_files = []  # Almacenar solo im√°genes con etiquetas v√°lidas

    for img_path in tqdm(image_files):
        base_name = os.path.basename(img_path)
        label_name = os.path.splitext(base_name)[0] + ".txt"
        label_path = os.path.join(src_label_dir, label_name)

        if os.path.exists(label_path):
            try:
                shutil.copy(img_path, os.path.join(output_image_dir, base_name))
                shutil.copy(label_path, os.path.join(output_label_dir, label_name))
                valid_image_files.append(img_path)  # A√±adir a lista de im√°genes v√°lidas
            except Exception as e:
                print(f"Error al copiar archivos originales: {e}")

    print(f"Total de im√°genes v√°lidas con etiquetas: {len(valid_image_files)}")

    # Aleatorizar im√°genes
    random.shuffle(valid_image_files)

    # Dividir en dos grupos para asegurar que no haya im√°genes repetidas
    images_per_type = total_images // 2

    # Garantizar que no intentamos procesar m√°s im√°genes de las disponibles
    if images_per_type * 2 > len(valid_image_files):
        print(f"ADVERTENCIA: Solo hay {len(valid_image_files)} im√°genes disponibles.")
        print(f"Reduciendo a {len(valid_image_files) // 2} im√°genes por tipo de aumentaci√≥n.")
        images_per_type = len(valid_image_files) // 2

    # Separar en dos grupos disjuntos (sin repetici√≥n)
    type1_images = valid_image_files[:images_per_type]
    type2_images = valid_image_files[images_per_type:images_per_type*2]

    print(f"Procesando {len(type1_images)} im√°genes para aumentaci√≥n tipo 1 (flip + brillo)")
    print(f"Procesando {len(type2_images)} im√°genes para aumentaci√≥n tipo 2 (rotaci√≥n)")

    # Verificar que no hay solapamiento
    common_images = set(type1_images).intersection(set(type2_images))
    if common_images:
        print(f"ERROR: Se detectaron {len(common_images)} im√°genes duplicadas entre ambos tipos.")
        print("Esto no deber√≠a ocurrir. Corrigiendo...")
        # No deber√≠a ocurrir con la l√≥gica actual, pero a√±adimos esta comprobaci√≥n por seguridad

    # Procesar aumentaci√≥n tipo 1: Flip horizontal con brillo
    print(f"Aplicando aumentaci√≥n tipo 1 (flip + brillo)...")
    type1_success = 0

    with tqdm(total=len(type1_images)) as pbar:
        for img_path in type1_images:
            base_name = os.path.basename(img_path)
            label_name = os.path.splitext(base_name)[0] + ".txt"
            label_path = os.path.join(src_label_dir, label_name)

            if os.path.exists(label_path):
                success = apply_augmentation_type(
                    img_path,
                    label_path,
                    output_image_dir,
                    output_label_dir,
                    transform_type1,
                    "flip_brightness",
                    visualize=visualize_sample and random.random() < 0.05
                )

                if success:
                    type1_success += 1
                pbar.update(1)

    # Procesar aumentaci√≥n tipo 2: Rotaci√≥n
    print(f"Aplicando aumentaci√≥n tipo 2 (rotaci√≥n)...")
    type2_success = 0

    with tqdm(total=len(type2_images)) as pbar:
        for img_path in type2_images:
            base_name = os.path.basename(img_path)
            label_name = os.path.splitext(base_name)[0] + ".txt"
            label_path = os.path.join(src_label_dir, label_name)

            if os.path.exists(label_path):
                success = apply_augmentation_type(
                    img_path,
                    label_path,
                    output_image_dir,
                    output_label_dir,
                    transform_type2,
                    "rotation",
                    visualize=visualize_sample and random.random() < 0.05
                )

                if success:
                    type2_success += 1
                pbar.update(1)

    print(f"Aumentaci√≥n completada para {split_name}:")
    print(f"  - Tipo 1 (flip + brillo): {type1_success} im√°genes procesadas exitosamente")
    print(f"  - Tipo 2 (rotaci√≥n): {type2_success} im√°genes procesadas exitosamente")
    print(f"  - Total: {type1_success + type2_success} im√°genes aumentadas")

def update_yaml_file():
    """Actualiza o crea el archivo data.yaml para el dataset aumentado"""
    src_yaml = os.path.join(DATASET_ROOT, "data.yaml")
    dst_yaml = os.path.join(OUTPUT_ROOT, "data.yaml")

    if os.path.exists(src_yaml):
        # Leer el YAML original
        with open(src_yaml, 'r') as f:
            yaml_content = f.read()

        # Actualizar rutas (para el caso de Drive, mejor usar rutas absolutas)
        updated_content = []
        for line in yaml_content.split('\n'):
            if line.startswith('train:') or line.startswith('val:') or line.startswith('test:'):
                key = line.split(':')[0].strip()
                updated_content.append(f"{key}: {os.path.join(OUTPUT_ROOT, 'images', key)}")
            else:
                updated_content.append(line)

        yaml_content = '\n'.join(updated_content)

        # Guardar nuevo YAML
        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml actualizado en {dst_yaml}")
    else:
        # Crear un YAML b√°sico
        train_path = os.path.join(OUTPUT_ROOT, "images", "train")
        val_path = os.path.join(OUTPUT_ROOT, "images", "val")
        test_path = os.path.join(OUTPUT_ROOT, "images", "test")

        yaml_content = f"""# Dataset configuraci√≥n para YOLOv11
train: {train_path}
val: {val_path}
test: {test_path}

# N√∫mero de clases
nc: 1

# Nombres de clases
names: ['person']
"""

        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml creado en {dst_yaml}")

def print_summary():
    """Imprime un resumen del proceso de aumentaci√≥n"""
    print("\n\n" + "="*60)
    print("RESUMEN DE AUMENTACI√ìN DE DATOS".center(60))
    print("="*60)

    for split in ["train", "val", "test"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        aug_img_dir = os.path.join(OUTPUT_ROOT, "images", split)

        if not os.path.exists(img_dir) or not os.path.exists(aug_img_dir):
            print(f"Split {split.upper()}: Directorio no encontrado, omitiendo...")
            continue

        orig_count = len(glob.glob(os.path.join(img_dir, "*.jpg")))
        if orig_count == 0:
            orig_count = len(glob.glob(os.path.join(img_dir, "*.png")))

        aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.jpg")))
        if aug_count == 0:
            aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.png")))

        # Contar por tipo de aumentaci√≥n
        aug_type1_count = len(glob.glob(os.path.join(aug_img_dir, "*_aug_flip_brightness.jpg")))
        aug_type2_count = len(glob.glob(os.path.join(aug_img_dir, "*_aug_rotation.jpg")))

        if orig_count > 0:
            print(f"Split {split.upper()}:")
            print(f"  - Im√°genes originales: {orig_count}")
            print(f"  - Aumentaci√≥n tipo 1 (flip + brillo): {aug_type1_count}")
            print(f"  - Aumentaci√≥n tipo 2 (rotaci√≥n): {aug_type2_count}")
            print(f"  - Total despu√©s de aumentaci√≥n: {aug_count}")
            print(f"  - Incremento: {aug_count - orig_count} im√°genes ({(aug_count/orig_count - 1)*100:.1f}% m√°s)")
            print("-"*60)

    print("\nVisualizaciones guardadas en:", VIZ_DIR)
    print(f"\nDataset aumentado guardado en: {OUTPUT_ROOT}")
    print("="*60)

def test_run_single_image():
    """Ejecuta una prueba en una sola imagen para verificar que todo funciona"""
    print("\nEjecutando prueba en una sola imagen...")

    # Buscar una imagen de prueba
    for split in ["train", "val"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        if os.path.exists(img_dir):
            image_files = glob.glob(os.path.join(img_dir, "*.jpg"))
            if not image_files:
                image_files = glob.glob(os.path.join(img_dir, "*.png"))

            if image_files:
                test_img = image_files[0]
                base_name = os.path.basename(test_img)
                label_name = os.path.splitext(base_name)[0] + ".txt"
                label_path = os.path.join(DATASET_ROOT, "labels", split, label_name)

                if os.path.exists(label_path):
                    print(f"Probando con imagen: {test_img}")
                    output_img_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    output_label_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    os.makedirs(output_img_dir, exist_ok=True)
                    os.makedirs(output_label_dir, exist_ok=True)

                    # Aplicar las dos aumentaciones
                    apply_augmentation_type(
                        test_img,
                        label_path,
                        output_img_dir,
                        output_label_dir,
                        transform_type1,
                        "flip_brightness",
                        visualize=True
                    )

                    apply_augmentation_type(
                        test_img,
                        label_path,
                        output_img_dir,
                        output_label_dir,
                        transform_type2,
                        "rotation",
                        visualize=True
                    )

                    print(f"Prueba completada. Verifica el directorio: {output_img_dir}")
                    return True

    print("No se pudo encontrar una imagen de prueba v√°lida.")
    return False
```

```python
import yaml
from google.colab import drive
import os
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
import cv2
import albumentations as A
import glob
import shutil
import random
import torch
import time

# Verificar disponibilidad de GPU
if torch.cuda.is_available():
    print(f"GPU disponible: {torch.cuda.get_device_name(0)}")
    print(f"Memoria GPU total: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
    print(f"Memoria GPU disponible: {torch.cuda.memory_allocated(0) / 1e9:.2f} GB utilizada")
    USE_GPU = True
else:
    print("GPU no disponible, utilizando CPU.")
    USE_GPU = False

# Montar Google Drive si a√∫n no est√° montado
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')
    print("Google Drive montado correctamente.")
else:
    print("Google Drive ya est√° montado.")

# DRIVE_BASE = "/content/drive/MyDrive/Proyecto-IA3"
DRIVE_BASE = "/content/drive/MyDrive/Proyecto IA-3/violence_school_project"
# DATASET_ROOT = f"/content/drive/MyDrive/dataset_people"
DATASET_ROOT = f"{DRIVE_BASE}/dataset_people"
OUTPUT_ROOT = f"{DRIVE_BASE}/dataset_people_augmented"  # Aqu√≠ se guardar√°n los datos aumentados

print(f"Ruta del dataset original: {DATASET_ROOT}")
print(f"Ruta donde se guardar√°n los datos aumentados: {OUTPUT_ROOT}")

# Crear directorios de salida si no existen
for split in ["train", "val", "test"]:
    for subdir in ["images", "labels"]:
        os.makedirs(os.path.join(OUTPUT_ROOT, subdir, split), exist_ok=True)

# Directorio para visualizaciones
VIZ_DIR = os.path.join(OUTPUT_ROOT, "visualizaciones")
os.makedirs(VIZ_DIR, exist_ok=True)

# Comprobar si las rutas existen
for split in ["train", "val", "test"]:
    img_dir = os.path.join(DATASET_ROOT, "images", split)
    label_dir = os.path.join(DATASET_ROOT, "labels", split)

    if os.path.exists(img_dir) and os.path.exists(label_dir):
        print(f"Directorios para {split} encontrados correctamente.")
    else:
        print(f"ADVERTENCIA: Los directorios para {split} no existen: {img_dir} o {label_dir}")

# Definimos exactamente dos transformaciones espec√≠ficas basadas en las que solicitaste

# Aumentaci√≥n tipo 1: Flip horizontal con m√°s brillo
transform_type1 = A.Compose([
    A.HorizontalFlip(p=1.0),  # Siempre hacer flip horizontal
    A.RandomBrightnessContrast(brightness_limit=(0.1, 0.2), contrast_limit=0, p=1.0),  # Aumentar brillo
], bbox_params=A.BboxParams(
    format='yolo',
    min_area=0.0001,
    min_visibility=0.3,
    label_fields=['class_labels']
))

# Aumentaci√≥n tipo 2: Peque√±a rotaci√≥n (como en la imagen 3)
transform_type2 = A.Compose([
    A.Rotate(limit=3, p=1.0),  # Rotaci√≥n muy sutil, siempre aplicada
], bbox_params=A.BboxParams(
    format='yolo',
    min_area=0.0001,
    min_visibility=0.3,
    label_fields=['class_labels']
))

# Configurar OpenCV para usar GPU si est√° disponible
if USE_GPU:
    # Intentar utilizar GPU para OpenCV - esto solo funciona si OpenCV est√° compilado con soporte CUDA
    try:
        cv2.cuda.setDevice(0)
        print("OpenCV configurado para usar GPU para aceleraci√≥n.")
    except:
        print("OpenCV no puede usar GPU directamente, utilizando aceleraci√≥n parcial.")

def visualize_bbox(img, bboxes, class_ids, class_name="person"):
    """Visualiza la imagen con sus bounding boxes para verificaci√≥n"""
    img = img.copy()
    h, w, _ = img.shape

    for i, bbox in enumerate(bboxes):
        # Convertir de formato YOLO [x_center, y_center, width, height] a p√≠xeles
        x_center, y_center, width, height = bbox
        x_min = int((x_center - width/2) * w)
        y_min = int((y_center - height/2) * h)
        x_max = int((x_center + width/2) * w)
        y_max = int((y_center + height/2) * h)

        # Dibujar rect√°ngulo
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

        # Etiqueta
        cv2.putText(img, class_name, (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    return img

# Procesamiento por lotes para mejor rendimiento en GPU
def batch_process_images(image_paths, label_paths, output_image_dir, output_label_dir,
                        transform_type, aug_name, visualize_prob=0.05, batch_size=16):
    """Procesa m√∫ltiples im√°genes en lotes para mejor rendimiento"""

    successful_augmentations = 0

    for batch_start in range(0, len(image_paths), batch_size):
        batch_end = min(batch_start + batch_size, len(image_paths))
        batch_img_paths = image_paths[batch_start:batch_end]
        batch_label_paths = label_paths[batch_start:batch_end]

        for i, (img_path, label_path) in enumerate(zip(batch_img_paths, batch_label_paths)):
            # Verificar que ambos archivos existen
            if not os.path.exists(img_path) or not os.path.exists(label_path):
                continue

            # Aplicar aumentaci√≥n
            success = apply_augmentation_type(
                img_path,
                label_path,
                output_image_dir,
                output_label_dir,
                transform_type,
                aug_name,
                visualize=random.random() < visualize_prob
            )

            if success:
                successful_augmentations += 1

    return successful_augmentations

def apply_augmentation_type(image_path, label_path, output_image_dir, output_label_dir,
                            transform_type, aug_name, visualize=False):
    """Aplica un tipo espec√≠fico de transformaci√≥n a una imagen"""
    # Leer imagen
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error al leer la imagen: {image_path}")
        return False

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Leer etiquetas
    try:
        with open(label_path, 'r') as f:
            lines = f.readlines()
    except Exception as e:
        print(f"Error al leer el archivo de etiquetas {label_path}: {e}")
        return False

    bboxes = []
    class_labels = []

    # Procesar etiquetas
    for line in lines:
        try:
            values = line.strip().split()
            class_id = int(values[0])
            x_center, y_center, width, height = [float(x) for x in values[1:5]]

            # Verificar si los valores est√°n en el rango correcto (0-1 para YOLO)
            if not all(0 <= val <= 1 for val in [x_center, y_center, width, height]):
                print(f"Advertencia: Valores fuera de rango en {label_path}: {values}")
                continue

            bboxes.append([x_center, y_center, width, height])
            class_labels.append(class_id)
        except Exception as e:
            print(f"Error al procesar l√≠nea en {label_path}: {line} - {e}")
            continue

    # Si no hay cajas v√°lidas, omitir esta imagen
    if not bboxes:
        print(f"No se encontraron bounding boxes v√°lidos en {label_path}")
        return False

    # Obtener nombre base del archivo
    base_name = os.path.basename(image_path).split('.')[0]

    try:
        # Aplicar transformaci√≥n (Albumentations ya utiliza aceleraci√≥n interna)
        augmented = transform_type(image=image, bboxes=bboxes, class_labels=class_labels)

        aug_image = augmented['image']
        aug_bboxes = augmented['bboxes']
        aug_class_labels = augmented['class_labels']

        # Verificar que la aumentaci√≥n contiene los bounding boxes esperados
        if not aug_bboxes:
            print(f"La transformaci√≥n {aug_name} elimin√≥ todos los bounding boxes para {image_path}, omitiendo.")
            return False

        # Guardar imagen aumentada
        output_image_path = os.path.join(output_image_dir, f"{base_name}_aug_{aug_name}.jpg")
        cv2.imwrite(output_image_path, cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR))

        # Guardar etiquetas aumentadas
        output_label_path = os.path.join(output_label_dir, f"{base_name}_aug_{aug_name}.txt")
        with open(output_label_path, 'w') as f:
            for bbox, class_id in zip(aug_bboxes, aug_class_labels):
                # Convertir class_id a entero antes de escribirlo
                class_id_int = int(class_id)
                f.write(f"{class_id_int} {' '.join([str(x) for x in bbox])}\n")

        # Visualizar aumentaci√≥n si se solicita
        if visualize:
            vis_aug_img = visualize_bbox(aug_image, aug_bboxes, aug_class_labels)
            plt.figure(figsize=(10, 10))
            plt.imshow(vis_aug_img)
            plt.title(f"Augmented Image - {aug_name}")
            plt.axis('off')
            plt.savefig(os.path.join(VIZ_DIR, f"{base_name}_aug_{aug_name}.jpg"), dpi=100)
            plt.close()

        return True  # La aumentaci√≥n se realiz√≥ correctamente

    except Exception as e:
        print(f"Error en la aumentaci√≥n {aug_name} para {image_path}: {e}")
        return False

def process_limited_split(split_name, total_images, visualize_sample=True, batch_size=16):
    """
    Procesa una cantidad limitada de im√°genes para cada tipo de aumentaci√≥n.
    Garantiza que no se procese la misma imagen dos veces.
    Ya no copia las im√°genes originales, solo guarda las aumentadas.

    Args:
        split_name: Nombre del split ('train' o 'val')
        total_images: N√∫mero total de im√°genes a procesar (se dividir√° entre los 2 tipos)
        visualize_sample: Si se deben visualizar algunas muestras
        batch_size: Tama√±o del lote para procesamiento en GPU
    """
    start_time = time.time()

    print(f"\n{'='*50}")
    print(f"Procesando split {split_name} con {total_images} im√°genes totales")
    print(f"{'='*50}")

    # Directorios originales
    src_image_dir = os.path.join(DATASET_ROOT, "images", split_name)
    src_label_dir = os.path.join(DATASET_ROOT, "labels", split_name)

    # Directorios de salida
    output_image_dir = os.path.join(OUTPUT_ROOT, "images", split_name)
    output_label_dir = os.path.join(OUTPUT_ROOT, "labels", split_name)

    # Obtener todas las im√°genes
    image_files = glob.glob(os.path.join(src_image_dir, "*.jpg"))
    if not image_files:
        image_files = glob.glob(os.path.join(src_image_dir, "*.png"))

    if not image_files:
        print(f"ADVERTENCIA: No se encontraron im√°genes en {src_image_dir}")
        return

    print(f"Encontradas {len(image_files)} im√°genes en {src_image_dir}")

    # Ya no copiamos archivos originales, solo guardamos los aumentados
    valid_image_files = []  # Almacenar solo im√°genes con etiquetas v√°lidas
    valid_label_files = []  # Almacenar las etiquetas correspondientes

    # Verificar qu√© im√°genes tienen etiquetas v√°lidas
    for img_path in tqdm(image_files, desc="Verificando im√°genes con etiquetas v√°lidas"):
        base_name = os.path.basename(img_path)
        label_name = os.path.splitext(base_name)[0] + ".txt"
        label_path = os.path.join(src_label_dir, label_name)

        if os.path.exists(label_path):
            valid_image_files.append(img_path)
            valid_label_files.append(label_path)

    print(f"Total de im√°genes v√°lidas con etiquetas: {len(valid_image_files)}")

    # Aleatorizar im√°genes pero mantener correspondencia con etiquetas
    combined = list(zip(valid_image_files, valid_label_files))
    random.shuffle(combined)
    valid_image_files, valid_label_files = zip(*combined)

    # Dividir en dos grupos para asegurar que no haya im√°genes repetidas
    images_per_type = total_images // 2

    # Garantizar que no intentamos procesar m√°s im√°genes de las disponibles
    if images_per_type * 2 > len(valid_image_files):
        print(f"ADVERTENCIA: Solo hay {len(valid_image_files)} im√°genes disponibles.")
        print(f"Reduciendo a {len(valid_image_files) // 2} im√°genes por tipo de aumentaci√≥n.")
        images_per_type = len(valid_image_files) // 2

    # Separar en dos grupos disjuntos (sin repetici√≥n)
    type1_images = valid_image_files[:images_per_type]
    type1_labels = valid_label_files[:images_per_type]

    type2_images = valid_image_files[images_per_type:images_per_type*2]
    type2_labels = valid_label_files[images_per_type:images_per_type*2]

    print(f"Procesando {len(type1_images)} im√°genes para aumentaci√≥n tipo 1 (flip + brillo)")
    print(f"Procesando {len(type2_images)} im√°genes para aumentaci√≥n tipo 2 (rotaci√≥n)")

    # Verificar que no hay solapamiento
    common_images = set(type1_images).intersection(set(type2_images))
    if common_images:
        print(f"ERROR: Se detectaron {len(common_images)} im√°genes duplicadas entre ambos tipos.")
        print("Esto no deber√≠a ocurrir. Corrigiendo...")

    # Procesar aumentaci√≥n tipo 1: Flip horizontal con brillo
    print(f"Aplicando aumentaci√≥n tipo 1 (flip + brillo) usando lotes de {batch_size} im√°genes...")
    if USE_GPU:
        # Procesamiento por lotes para GPU
        successful_type1 = batch_process_images(
            type1_images,
            type1_labels,
            output_image_dir,
            output_label_dir,
            transform_type1,
            "flip_brightness",
            visualize_prob=0.05 if visualize_sample else 0,
            batch_size=batch_size
        )
    else:
        # Procesamiento secuencial para CPU
        successful_type1 = 0
        with tqdm(total=len(type1_images)) as pbar:
            for img_path, label_path in zip(type1_images, type1_labels):
                success = apply_augmentation_type(
                    img_path,
                    label_path,
                    output_image_dir,
                    output_label_dir,
                    transform_type1,
                    "flip_brightness",
                    visualize=visualize_sample and random.random() < 0.05
                )

                if success:
                    successful_type1 += 1
                pbar.update(1)

    # Procesar aumentaci√≥n tipo 2: Rotaci√≥n
    print(f"Aplicando aumentaci√≥n tipo 2 (rotaci√≥n) usando lotes de {batch_size} im√°genes...")
    if USE_GPU:
        # Procesamiento por lotes para GPU
        successful_type2 = batch_process_images(
            type2_images,
            type2_labels,
            output_image_dir,
            output_label_dir,
            transform_type2,
            "rotation",
            visualize_prob=0.05 if visualize_sample else 0,
            batch_size=batch_size
        )
    else:
        # Procesamiento secuencial para CPU
        successful_type2 = 0
        with tqdm(total=len(type2_images)) as pbar:
            for img_path, label_path in zip(type2_images, type2_labels):
                success = apply_augmentation_type(
                    img_path,
                    label_path,
                    output_image_dir,
                    output_label_dir,
                    transform_type2,
                    "rotation",
                    visualize=visualize_sample and random.random() < 0.05
                )

                if success:
                    successful_type2 += 1
                pbar.update(1)

    end_time = time.time()
    total_time = end_time - start_time

    print(f"Aumentaci√≥n completada para {split_name} en {total_time:.2f} segundos:")
    print(f"  - Tipo 1 (flip + brillo): {successful_type1} im√°genes procesadas exitosamente")
    print(f"  - Tipo 2 (rotaci√≥n): {successful_type2} im√°genes procesadas exitosamente")
    print(f"  - Total: {successful_type1 + successful_type2} im√°genes aumentadas")
    print(f"  - Velocidad: {(successful_type1 + successful_type2) / total_time:.2f} im√°genes/segundo")

def update_yaml_file():
    """Actualiza o crea el archivo data.yaml para el dataset aumentado"""
    src_yaml = os.path.join(DATASET_ROOT, "data.yaml")
    dst_yaml = os.path.join(OUTPUT_ROOT, "data.yaml")

    if os.path.exists(src_yaml):
        # Leer el YAML original
        with open(src_yaml, 'r') as f:
            yaml_content = f.read()

        # Actualizar rutas (para el caso de Drive, mejor usar rutas absolutas)
        updated_content = []
        for line in yaml_content.split('\n'):
            if line.startswith('train:') or line.startswith('val:') or line.startswith('test:'):
                key = line.split(':')[0].strip()
                updated_content.append(f"{key}: {os.path.join(OUTPUT_ROOT, 'images', key)}")
            else:
                updated_content.append(line)

        yaml_content = '\n'.join(updated_content)

        # Guardar nuevo YAML
        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml actualizado en {dst_yaml}")
    else:
        # Crear un YAML b√°sico
        train_path = os.path.join(OUTPUT_ROOT, "images", "train")
        val_path = os.path.join(OUTPUT_ROOT, "images", "val")
        test_path = os.path.join(OUTPUT_ROOT, "images", "test")

        yaml_content = f"""# Dataset configuraci√≥n para YOLOv11
train: {train_path}
val: {val_path}
test: {test_path}

# N√∫mero de clases
nc: 1

# Nombres de clases
names: ['person']
"""

        with open(dst_yaml, 'w') as f:
            f.write(yaml_content)

        print(f"Archivo data.yaml creado en {dst_yaml}")

def print_summary():
    """Imprime un resumen del proceso de aumentaci√≥n"""
    print("\n\n" + "="*60)
    print("RESUMEN DE AUMENTACI√ìN DE DATOS".center(60))
    print("="*60)

    for split in ["train", "val", "test"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        aug_img_dir = os.path.join(OUTPUT_ROOT, "images", split)

        if not os.path.exists(img_dir) or not os.path.exists(aug_img_dir):
            print(f"Split {split.upper()}: Directorio no encontrado, omitiendo...")
            continue

        orig_count = len(glob.glob(os.path.join(img_dir, "*.jpg")))
        if orig_count == 0:
            orig_count = len(glob.glob(os.path.join(img_dir, "*.png")))

        aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.jpg")))
        if aug_count == 0:
            aug_count = len(glob.glob(os.path.join(aug_img_dir, "*.png")))

        # Contar por tipo de aumentaci√≥n
        aug_type1_count = len(glob.glob(os.path.join(aug_img_dir, "*_aug_flip_brightness.jpg")))
        aug_type2_count = len(glob.glob(os.path.join(aug_img_dir, "*_aug_rotation.jpg")))

        if orig_count > 0:
            print(f"Split {split.upper()}:")
            print(f"  - Im√°genes en el dataset original: {orig_count}")
            print(f"  - Aumentaci√≥n tipo 1 (flip + brillo): {aug_type1_count}")
            print(f"  - Aumentaci√≥n tipo 2 (rotaci√≥n): {aug_type2_count}")
            print(f"  - Total im√°genes aumentadas: {aug_count}")
            print("-"*60)

    print("\nVisualizaciones guardadas en:", VIZ_DIR)
    print(f"\nDataset aumentado guardado en: {OUTPUT_ROOT}")
    print("="*60)

def test_run_single_image():
    """Ejecuta una prueba en una sola imagen para verificar que todo funciona"""
    print("\nEjecutando prueba en una sola imagen...")

    # Buscar una imagen de prueba
    for split in ["train", "val"]:
        img_dir = os.path.join(DATASET_ROOT, "images", split)
        if os.path.exists(img_dir):
            image_files = glob.glob(os.path.join(img_dir, "*.jpg"))
            if not image_files:
                image_files = glob.glob(os.path.join(img_dir, "*.png"))

            if image_files:
                test_img = image_files[0]
                base_name = os.path.basename(test_img)
                label_name = os.path.splitext(base_name)[0] + ".txt"
                label_path = os.path.join(DATASET_ROOT, "labels", split, label_name)

                if os.path.exists(label_path):
                    print(f"Probando con imagen: {test_img}")
                    output_img_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    output_label_dir = os.path.join(OUTPUT_ROOT, "test_single")
                    os.makedirs(output_img_dir, exist_ok=True)
                    os.makedirs(output_label_dir, exist_ok=True)

                    # Aplicar las dos aumentaciones
                    apply_augmentation_type(
                        test_img,
                        label_path,
                        output_img_dir,
                        output_label_dir,
                        transform_type1,
                        "flip_brightness",
                        visualize=True
                    )

                    apply_augmentation_type(
                        test_img,
                        label_path,
                        output_img_dir,
                        output_label_dir,
                        transform_type2,
                        "rotation",
                        visualize=True
                    )

                    print(f"Prueba completada. Verifica el directorio: {output_img_dir}")
                    return True

    print("No se pudo encontrar una imagen de prueba v√°lida.")
    return False
```

```python
import os
import glob
from tqdm import tqdm

# Ruta al dataset aumentado
DRIVE_BASE = "/content/drive/MyDrive/Proyecto-IA3"
OUTPUT_ROOT = f"{DRIVE_BASE}/dataset_people_augmented"

def fix_label_files():
    """Corrige los archivos de etiquetas para asegurar que el ID de clase sea un entero"""
    print("Corrigiendo archivos de etiquetas...")

    # Buscar todos los archivos de etiquetas aumentados
    label_files = []
    for split in ["train", "val"]:
        label_dir = os.path.join(OUTPUT_ROOT, "labels", split)
        if os.path.exists(label_dir):
            # Buscar solo archivos que contienen "_aug_" en el nombre
            aug_files = glob.glob(os.path.join(label_dir, "*_aug_*.txt"))
            label_files.extend(aug_files)

    print(f"Encontrados {len(label_files)} archivos de etiquetas a corregir.")

    # Corregir cada archivo
    fixed_count = 0
    for label_path in tqdm(label_files):
        try:
            # Leer el contenido actual
            with open(label_path, 'r') as f:
                lines = f.readlines()

            # Corregir cada l√≠nea
            new_lines = []
            modified = False

            for line in lines:
                parts = line.strip().split()
                if len(parts) >= 5:
                    # Verificar si el primer valor es un n√∫mero flotante con punto decimal
                    if '.' in parts[0]:
                        try:
                            class_id = int(float(parts[0]))  # Convertir a entero
                            # Reconstruir la l√≠nea con el ID de clase como entero
                            new_line = f"{class_id} {' '.join(parts[1:])}\n"
                            new_lines.append(new_line)
                            modified = True
                            continue
                        except ValueError:
                            pass

                # Si no necesita correcci√≥n o hubo un error, mantener la l√≠nea original
                new_lines.append(line)

            # Guardar el archivo corregido si se realizaron cambios
            if modified:
                with open(label_path, 'w') as f:
                    f.writelines(new_lines)
                fixed_count += 1

        except Exception as e:
            print(f"Error al procesar archivo {label_path}: {e}")

    print(f"Corregidos {fixed_count} archivos de etiquetas.")

# Ejecutar la correcci√≥n
fix_label_files()
```

```python
import os
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import glob
from tqdm import tqdm

# Configuraci√≥n de rutas
DRIVE_BASE = "/content/drive/MyDrive/Proyecto IA-3/violence_school_project"
AUGMENTED_ROOT = f"{DRIVE_BASE}/dataset_people_augmented"

# Directorio para guardar las visualizaciones si quieres
VIZ_SAMPLES_DIR = os.path.join(AUGMENTED_ROOT, "visualizaciones_muestras")
os.makedirs(VIZ_SAMPLES_DIR, exist_ok=True)

# N√∫mero de muestras aleatorias a visualizar
NUM_SAMPLES = 50

# Definimos una paleta de colores para cada tipo de aumentaci√≥n
COLORS = {
    "flip_brightness": (0, 255, 0),  # Verde para flip + brillo
    "rotation": (255, 0, 0),        # Rojo para rotaci√≥n
}

def visualize_bbox(img, bboxes, labels, color=(0, 255, 0), thickness=2):
    """Visualiza la imagen con sus bounding boxes"""
    h, w, _ = img.shape

    for i, bbox in enumerate(bboxes):
        # Convertir de formato YOLO [x_center, y_center, width, height] a p√≠xeles
        x_center, y_center, width, height = bbox
        x_min = int((x_center - width/2) * w)
        y_min = int((y_center - height/2) * h)
        x_max = int((x_center + width/2) * w)
        y_max = int((y_center + height/2) * h)

        # Restricci√≥n para evitar valores fuera de la imagen
        x_min = max(0, x_min)
        y_min = max(0, y_min)
        x_max = min(w, x_max)
        y_max = min(h, y_max)

        # Dibujar rect√°ngulo
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, thickness)

        # Etiqueta
        label_text = f"person {i+1}"
        cv2.putText(img, label_text, (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    return img

def get_augmentation_type(filename):
    """Determina el tipo de aumentaci√≥n basado en el nombre del archivo"""
    if "_aug_flip_brightness." in filename:
        return "flip_brightness"
    elif "_aug_rotation." in filename:
        return "rotation"
    else:
        return "original"

def load_label(label_path):
    """Carga las etiquetas desde un archivo .txt"""
    bboxes = []
    labels = []

    try:
        with open(label_path, 'r') as f:
            lines = f.readlines()

        for line in lines:
            values = line.strip().split()
            class_id = int(values[0])
            x_center, y_center, width, height = [float(x) for x in values[1:5]]

            # Verificar si los valores est√°n en el rango correcto
            if 0 <= x_center <= 1 and 0 <= y_center <= 1 and 0 <= width <= 1 and 0 <= height <= 1:
                bboxes.append([x_center, y_center, width, height])
                labels.append(class_id)
            else:
                print(f"Advertencia: Valores fuera de rango en {label_path}: {values}")
    except Exception as e:
        print(f"Error al leer el archivo de etiquetas {label_path}: {e}")

    return bboxes, labels

def get_all_augmented_images():
    """Obtiene todas las im√°genes aumentadas del dataset"""
    augmented_images = []

    # Buscar en train y val
    for split in ["train", "val"]:
        image_dir = os.path.join(AUGMENTED_ROOT, "images", split)

        if os.path.exists(image_dir):
            # Buscar im√°genes aumentadas (contienen _aug_ en el nombre)
            aug_files = glob.glob(os.path.join(image_dir, "*_aug_*.jpg"))

            for img_path in aug_files:
                label_name = os.path.basename(img_path).replace('.jpg', '.txt')
                label_path = os.path.join(AUGMENTED_ROOT, "labels", split, label_name)

                if os.path.exists(label_path):
                    augmented_images.append((img_path, label_path, split))

    return augmented_images

def visualize_random_samples(num_samples=50, save_to_disk=True, fig_size=(15, 15), cols=5):
    """Visualiza un n√∫mero espec√≠fico de muestras aleatorias del dataset aumentado"""
    # Obtener todas las im√°genes aumentadas
    all_images = get_all_augmented_images()
    print(f"Total de im√°genes aumentadas encontradas: {len(all_images)}")

    # Seleccionar muestras aleatorias
    if num_samples > len(all_images):
        num_samples = len(all_images)
        print(f"Ajustando a {num_samples} muestras disponibles")

    samples = random.sample(all_images, num_samples)

    # Calcular n√∫mero de filas necesarias
    rows = (num_samples + cols - 1) // cols

    # Crear figura grande
    plt.figure(figsize=fig_size)

    for i, (img_path, label_path, split) in enumerate(tqdm(samples, desc="Visualizando muestras")):
        # Leer imagen
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error al leer la imagen: {img_path}")
            continue

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Leer etiquetas
        bboxes, labels = load_label(label_path)

        # Determinar tipo de aumentaci√≥n
        aug_type = get_augmentation_type(img_path)
        color = COLORS.get(aug_type, (0, 0, 255))  # Color predeterminado azul

        # Visualizar bounding boxes
        img_with_boxes = visualize_bbox(img.copy(), bboxes, labels, color=color)

        # A√±adir a la figura
        plt.subplot(rows, cols, i+1)
        plt.imshow(img_with_boxes)
        plt.title(f"{split} - {aug_type}")
        plt.axis('off')

    plt.tight_layout()

    # Guardar visualizaci√≥n si se solicita
    if save_to_disk:
        output_path = os.path.join(VIZ_SAMPLES_DIR, f"random_samples_{num_samples}.jpg")
        plt.savefig(output_path, dpi=150, bbox_inches='tight')
        print(f"Visualizaci√≥n guardada en: {output_path}")

    plt.show()

# Funci√≥n para visualizar una sola imagen con sus bounding boxes
def visualize_single_image(img_path, save_to_disk=False):
    """Visualiza una sola imagen con sus bounding boxes"""
    split = "train" if "train" in img_path else "val" if "val" in img_path else "test"

    # Determinar ruta del label
    label_name = os.path.basename(img_path).replace('.jpg', '.txt')
    label_path = os.path.join(AUGMENTED_ROOT, "labels", split, label_name)

    if not os.path.exists(label_path):
        print(f"No se encontr√≥ el archivo de etiquetas: {label_path}")
        return

    # Leer imagen
    img = cv2.imread(img_path)
    if img is None:
        print(f"Error al leer la imagen: {img_path}")
        return

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Leer etiquetas
    bboxes, labels = load_label(label_path)

    # Determinar tipo de aumentaci√≥n
    aug_type = get_augmentation_type(img_path)
    color = COLORS.get(aug_type, (0, 0, 255))  # Color predeterminado azul

    # Visualizar bounding boxes
    img_with_boxes = visualize_bbox(img.copy(), bboxes, labels, color=color)

    # Mostrar imagen
    plt.figure(figsize=(10, 10))
    plt.imshow(img_with_boxes)
    plt.title(f"{split} - {aug_type} - {os.path.basename(img_path)}")
    plt.axis('off')

    # Guardar visualizaci√≥n si se solicita
    if save_to_disk:
        output_path = os.path.join(VIZ_SAMPLES_DIR, f"single_{os.path.basename(img_path)}")
        plt.savefig(output_path, dpi=150, bbox_inches='tight')
        print(f"Visualizaci√≥n guardada en: {output_path}")

    plt.show()
```

```python
import os
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import glob
from tqdm import tqdm

# Configuraci√≥n de rutas
DRIVE_BASE = "/content/drive/MyDrive/Proyecto IA-3/violence_school_project"
AUGMENTED_ROOT = f"{DRIVE_BASE}/dataset_people_augmented"

# Directorio para guardar las visualizaciones
VIZ_SAMPLES_DIR = os.path.join(AUGMENTED_ROOT, "visualizaciones_muestras")
os.makedirs(VIZ_SAMPLES_DIR, exist_ok=True)

# N√∫mero de muestras aleatorias a visualizar
NUM_SAMPLES = 50

# Definimos una paleta de colores para cada tipo de aumentaci√≥n
COLORS = {
    "flip_brightness": (0, 255, 0),  # Verde para flip + brillo
    "rotation": (255, 0, 0),        # Rojo para rotaci√≥n
}

def visualize_bbox(img, bboxes, labels, color=(0, 255, 0), thickness=2):
    """Visualiza la imagen con sus bounding boxes"""
    h, w, _ = img.shape

    for i, bbox in enumerate(bboxes):
        # Convertir de formato YOLO [x_center, y_center, width, height] a p√≠xeles
        x_center, y_center, width, height = bbox
        x_min = int((x_center - width/2) * w)
        y_min = int((y_center - height/2) * h)
        x_max = int((x_center + width/2) * w)
        y_max = int((y_center + height/2) * h)

        # Restricci√≥n para evitar valores fuera de la imagen
        x_min = max(0, min(w-1, x_min))
        y_min = max(0, min(h-1, y_min))
        x_max = max(0, min(w, x_max))
        y_max = max(0, min(h, y_max))

        # Dibujar rect√°ngulo
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color, thickness)

        # Etiqueta
        label_text = f"person {i+1}"
        cv2.putText(img, label_text, (x_min, y_min - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    return img

def get_augmentation_type(filename):
    """Determina el tipo de aumentaci√≥n basado en el nombre del archivo"""
    if "_aug_flip_brightness." in filename:
        return "flip_brightness"
    elif "_aug_rotation." in filename:
        return "rotation"
    else:
        return "original"

def load_label(label_path):
    """Carga las etiquetas desde un archivo .txt, manejando correctamente flotantes y enteros"""
    bboxes = []
    labels = []

    try:
        with open(label_path, 'r') as f:
            lines = f.readlines()

        for line in lines:
            values = line.strip().split()
            if len(values) >= 5:
                # Manejar tanto enteros como flotantes para el class_id
                try:
                    # Primero intentar como entero
                    class_id = int(values[0])
                except ValueError:
                    # Si falla, intentar como flotante
                    try:
                        class_id = int(float(values[0]))
                    except ValueError:
                        print(f"Advertencia: Valor de clase inv√°lido en {label_path}: {values[0]}")
                        continue

                # Procesar coordenadas
                try:
                    x_center = float(values[1])
                    y_center = float(values[2])
                    width = float(values[3])
                    height = float(values[4])

                    # Verificar rango
                    if 0 <= x_center <= 1 and 0 <= y_center <= 1 and 0 <= width <= 1 and 0 <= height <= 1:
                        bboxes.append([x_center, y_center, width, height])
                        labels.append(class_id)
                    else:
                        print(f"Advertencia: Coordenadas fuera de rango en {label_path}")
                except Exception as e:
                    print(f"Error al procesar coordenadas en {label_path}: {e}")
    except Exception as e:
        print(f"Error al leer el archivo de etiquetas {label_path}: {e}")

    return bboxes, labels

def get_all_augmented_images():
    """Obtiene todas las im√°genes aumentadas del dataset"""
    augmented_images = []

    # Buscar en train y val
    for split in ["train", "val"]:
        image_dir = os.path.join(AUGMENTED_ROOT, "images", split)

        if os.path.exists(image_dir):
            # Buscar im√°genes aumentadas (contienen _aug_ en el nombre)
            aug_files = glob.glob(os.path.join(image_dir, "*_aug_*.jpg"))

            for img_path in aug_files:
                label_name = os.path.basename(img_path).replace('.jpg', '.txt')
                label_path = os.path.join(AUGMENTED_ROOT, "labels", split, label_name)

                if os.path.exists(label_path):
                    augmented_images.append((img_path, label_path, split))

    return augmented_images

def visualize_random_samples(num_samples=50, save_to_disk=True, fig_size=(15, 15), cols=5):
    """Visualiza un n√∫mero espec√≠fico de muestras aleatorias del dataset aumentado"""
    # Obtener todas las im√°genes aumentadas
    all_images = get_all_augmented_images()
    print(f"Total de im√°genes aumentadas encontradas: {len(all_images)}")

    # Seleccionar muestras aleatorias
    if num_samples > len(all_images):
        num_samples = len(all_images)
        print(f"Ajustando a {num_samples} muestras disponibles")

    samples = random.sample(all_images, num_samples)

    # Calcular n√∫mero de filas necesarias
    rows = (num_samples + cols - 1) // cols

    # Crear figura grande
    plt.figure(figsize=fig_size)

    successful_samples = 0
    for i, (img_path, label_path, split) in enumerate(tqdm(samples, desc="Visualizando muestras")):
        try:
            # Leer imagen
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error al leer la imagen: {img_path}")
                continue

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Leer etiquetas con manejo mejorado
            bboxes, labels = load_label(label_path)

            if not bboxes:
                print(f"No se encontraron bounding boxes v√°lidos en {label_path}")
                continue

            # Determinar tipo de aumentaci√≥n
            aug_type = get_augmentation_type(img_path)
            color = COLORS.get(aug_type, (0, 0, 255))  # Color predeterminado azul

            # Visualizar bounding boxes
            img_with_boxes = visualize_bbox(img.copy(), bboxes, labels, color=color)

            # A√±adir a la figura
            plt.subplot(rows, cols, successful_samples+1)
            plt.imshow(img_with_boxes)
            plt.title(f"{split} - {aug_type}")
            plt.axis('off')

            successful_samples += 1

            # Si hemos completado suficientes muestras exitosas, terminar
            if successful_samples >= num_samples:
                break

        except Exception as e:
            print(f"Error al procesar {img_path}: {e}")

    plt.tight_layout()

    # Guardar visualizaci√≥n si se solicita
    if save_to_disk:
        output_path = os.path.join(VIZ_SAMPLES_DIR, f"random_samples_{successful_samples}.jpg")
        plt.savefig(output_path, dpi=150, bbox_inches='tight')
        print(f"Visualizaci√≥n guardada en: {output_path}")

    plt.show()

    return successful_samples

# Ejecutar la visualizaci√≥n
if __name__ == "__main__":
    # Visualizar 50 muestras aleatorias
    num_visualized = visualize_random_samples(NUM_SAMPLES, save_to_disk=True, fig_size=(20, 20), cols=5)
    print(f"Se visualizaron correctamente {num_visualized} im√°genes con sus bounding boxes")
```
